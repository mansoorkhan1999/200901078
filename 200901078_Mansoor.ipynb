```python

class Node:
  def __init__(self,value):
    self.value = value
    self.left = None
    self.right = None
    
     
class BTree:
  def __init__ (self, root):
    self.root = Node(root)
    
  
  #traversal algorithms:
  
  def preorderPrint(self, start, traversal): 
    '''root-left-right''' 
    if start:
      traversal += (str(start.value) )
      traversal = self.preorderPrint(start.left, traversal)
      traversal = self.preorderPrint(start.right, traversal)
    return traversal

  def inorderPrint(self, start, traversal):
    '''left-root-right'''
    if start:
      traversal = self.inorderPrint(start.left, traversal)
      traversal += (str(start.value) )
      traversal = self.inorderPrint(start.right, traversal)
    return traversal

  def postorderPrint(self, start, traversal):
    '''left-right-root'''
    if start:
      traversal = self.postorderPrint(start.left, traversal)
      traversal = self.postorderPrint(start.right, traversal)
      traversal += (str(start.value) )
    return traversal

  def print_traversal(self, traversal_type):
    if traversal_type =='Preorder':
      return self.preorderPrint(self.root, "") #provide start node as argument
    elif traversal_type == 'Inorder':
      return self.inorderPrint(self.root, "")
      #print(root.value)
    elif traversal_type == 'Postorder':
      return self.postorderPrint(self.root, "")
    else:
      print("Invalid Statement")
      return False

  def solve_eq(self, root, expression):
      if root:
        self.solve_eq(root.left, expression)
        self.solve_eq(root.right, expression)
        expression.append(root.value)
      return expression 
  
  def evaluate_expression(self, root):
      expression = []
      expression = self.solve_eq(root, expression)
      
      print("--------Original equation (in postfix)----------")
      for val in expression:
        print(val, end = "")
      print()

      answer = []
      for i in expression:
        if i in '0123456789':
          answer.append(i)
        else:
          #print(answer)
          op2 = answer.pop()
          op1 = answer.pop()
          if i == '*':
            operation = int(op1)*int(op2)
          elif i == '/':
            operation = int(op1)/int(op2)
          elif i == '+':
            operation = int(op1)+int(op2)
          elif i == '-':
            operation = int(op1)-int(op2)
          elif i == '^':
            operation = int(op1)^int(op2)
          answer.append(operation)
      print("----------Evaluated expression-----------")
      print(answer.pop())
      
def quiz_a():
    mytree = BTree('+') #root node
    
    #-----left subtree-----
    mytree.root.left = Node('/') #parent node
    
    mytree.root.left.left = Node('*')
    mytree.root.left.left.left = Node('0')
    mytree.root.left.left.right = Node('0')
    
    mytree.root.left.right = Node('-')
    mytree.root.left.right.left = Node('9')
    mytree.root.left.right.right = Node('0')
    
    #-----right subtree-----
    mytree.root.right = Node('*') #parent node
    
    mytree.root.right.left = Node('+')
    mytree.root.right.left.left = Node('1')
    mytree.root.right.left.right = Node('0')
    
    mytree.root.right.right = Node('/')
    mytree.root.right.right.left = Node('7')
    mytree.root.right.right.right = Node('8')
    print("Preorder : ")
    print(mytree.print_traversal('Preorder'))
    print("Inorder : ")
    print(mytree.print_traversal('Inorder'))
    print("Postorder : ")
    print(mytree.print_traversal('Postorder'))
    
    mytree.evaluate_expression(mytree.root)

     
def quiz_b():  #my attempted quiz which gave answer as infinity
    mytree = BTree('-') #root node
    
    #-----left subtree-----
    mytree.root.left = Node('*') #parent node
    
    mytree.root.left.left = Node('-')
    mytree.root.left.left.left = Node('0')
    mytree.root.left.left.right = Node('0')
    
    mytree.root.left.right = Node('/')
    mytree.root.left.right.left = Node('9')
    mytree.root.left.right.right = Node('0')
    
    #-----right subtree-----
    mytree.root.right = Node('/') #parent node
    
    mytree.root.right.left = Node('*')
    mytree.root.right.left.left = Node('1')
    mytree.root.right.left.right = Node('0')
    
    mytree.root.right.right = Node('+')
    mytree.root.right.right.left = Node('7')
    mytree.root.right.right.right = Node('8')
    print("Preorder : ")
    print(mytree.print_traversal('Preorder'))
    print("Inorder : ")
    print(mytree.print_traversal('Inorder'))
    print("Postorder : ")
    print(mytree.print_traversal('Postorder'))
    
    mytree.evaluate_expression(mytree.root)

def main_menu():
    selection = input("Which quiz you want to select: 'A' or 'B'?")
    if selection == 'A':
        print("You have selected Quiz 'A'  ")
        quiz_a()
    elif selection == 'B':
        print("You have selected Quiz 'B'  ")
        quiz_b()
    else:
        print("Incorrect selection! Try again. ")
        
main_menu()
```
